### Count or Frequency Encoding

```python
df = pd.read_csv('Datasets/adult.data' , header = None, index_col = None)
df.head()
```

```python
df.shape
```

```python
cols = [x for x in df.columns if df[x].dtypes == 'O']
cols.remove(14)
cols
```

```python
df[cols].head()
```

```python
df = df[cols]
df.head()
```

```python
df.columns=['Employment','Degree','Status','Designation','family_job','Race','Sex','Country']
df.head()
```

```python
for col in df.columns:
    print("For {}: we have {} categories".format(col, len(df[col].value_counts())))
```

```python
country_map=df['Country'].value_counts().to_dict()
country_map
```

```python
df['Country']=df['Country'].map(country_map)
df.head(20)
```

#### Advantages:

    - Easy To Use.
    - Not increasing feature space .

#### Disadvantages:
    - It will provide same weight if the frequencies are same

### Target Guided Ordinal Encoding:

    - Ordering the labels according to the target
    - Replace the labels by the joint probability of being 1 or 0

```python
import pandas as pd
df=pd.read_csv('Datasets/titanic_train.csv', usecols=['Cabin','Survived'])
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df['Cabin'].fillna('Missing',inplace=True)
df.head()
```

```python
df.isnull().sum()
```

```python
df.Cabin.value_counts()
```

```python
df['Cabin']=df['Cabin'].astype(str).str[0]
df.head()
```

```python
df.Cabin.unique()
```

```python
df.groupby(['Cabin'])['Survived'].mean()
```

```python
df.groupby(['Cabin'])['Survived'].mean().sort_values()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Cabin', hue='Survived')
plt.show()
```

```python
df.groupby(['Cabin'])['Survived'].mean().sort_values().index
```

```python
ordinal_labels=df.groupby(['Cabin'])['Survived'].mean().sort_values().index
ordinal_labels
```

```python
enumerate(ordinal_labels,0)
```

```python
ordinal_labels2 = {k:i for i, k in enumerate(ordinal_labels,0)}
ordinal_labels2
```

```python
df['Cabin_ordinal_labels']=df['Cabin'].map(ordinal_labels2)
df.head()
```

### Mean Encoding

```python
mean_ordinal=df.groupby(['Cabin'])['Survived'].mean().to_dict()
mean_ordinal
```

```python
df['mean_ordinal_encode']=df['Cabin'].map(mean_ordinal)
df.head()
```
